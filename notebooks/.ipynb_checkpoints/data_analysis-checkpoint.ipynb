# Correlation heatmap

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv(r'C:\Users\33670\Desktop\framework\data_collection\historical_data\TSLA_09082023.csv')

df = df[['first_hour_move','first_hour_max_up','first_hour_max_down','move_from_breakout_of_pm_high','week_day','gap','consecutive_days_above_20sma','open_from_BB_up','open_from_20sma','open_from_50sma','open_from_200sma','duration_before_breakout_of_pm_high','from_pm_high_to_open','percent_of_red_candles_one_hour_before_open','x_day_high_at_open','sigma_ratio_open_from_20sma','1month_change_from_high','red_days','green_days','lower_move_before_breakout_of_pm_high','days_til_next_earnings']]

# Calculez la matrice de corrélation
corr_matrix = df.corr()

# Créez une heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Heatmap de Corrélation des Caractéristiques")
plt.show()
```
